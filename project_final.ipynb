```python
c=0
import time
links=[]
ratings=[]
price=[]
for page in range(1,5):
    c+=1
    items= driver.find_elements(By.CLASS_NAME, 'gridItem--Yd0sa')
    for i in items:
        price_element = i.find_element(By.CLASS_NAME, 'currency--GVKjl')
        price_text = price_element.text
        amount_without_currency = price_text.replace('Rs', '').strip()
        clean_amount = amount_without_currency.replace(',', '')
        digits_only = ''.join(char for char in clean_amount if char.isdigit())
        price.append(digits_only)
        try:
            r = i.find_element(By.CLASS_NAME,'rating__review--ygkUy')
            ratings.append(r.text)
        except Exception as e:
            ratings.append('(0)')
        anchor_tags = i.find_elements(By.TAG_NAME, 'a')
        for tag in anchor_tags:
            download_link = tag.get_attribute("href")
            links.append( download_link )
            break
    if c==4:
        a=driver.find_element(By.XPATH,'/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li[5]')
        a.click()
    else:
        h=driver.find_element(By.XPATH,'/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[3]/div[2]/div/ul/li[9]/a')
        h.click()
    time.sleep(5)
```

```python
import pandas as pd

csv_file_path = 'project_daraz2.xlsx'

# Read the CSV file into a DataFrame
df = pd.read_excel(csv_file_path)
```

```python
df=pd.DataFrame(df)
```

```python
import re
ram_pattern = r'(\d+)GB'
camera_pattern = r'(\d+)\s*MP'
# Extract RAM and Camera information using regular expressions
df['RAM'] = df['Specification'].str.extract(ram_pattern, expand=False)
df['Camera'] = df['Specification'].str.extract(camera_pattern, expand=False)
# Convert extracted values to numeric (if needed)
df['RAM'] = pd.to_numeric(df['RAM'], errors='coerce')
df['Camera'] = pd.to_numeric(df['Camera'], errors='coerce')
```
